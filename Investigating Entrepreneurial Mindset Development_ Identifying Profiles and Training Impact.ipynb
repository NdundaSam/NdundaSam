```python
#Importing necessary libraries 
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#Importing dataset
file_path = r'C:\Users\Administrator\Documents\PYTHON WORK\Entreprenual Mindset\Test English (1).csv'
data = pd.read_csv(file_path)

#Summary statistics of the data
display(data.describe())

display(data.head(15))
```

```python
# independence: Measure of the degree of independence exhibited by respondents
# structure: Measure of the preference for structure and organization
# non-conformity: Measure of the tendency to deviate from conventional norms
# risk acceptance: Measure of the willingness to take risks
# Action orientation: Measure of the level of proactivity and action-taking
# Passion: Measure of the intensity of passion and enthusiasm
# Performance need: Measure of the drive for achievement and high performance
# Focus on the future: Measure of the emphasis on future-oriented thinking
# Idea generation: Measure of the ability to generate innovative ideas
# export: Measure of the interest and involvement in export activities
# Self confidence: Measure of self-assurance and belief in one's abilities
# Optimism: Measure of the level of optimism and positive outlook
# Perseverance: Measure of the ability to persist in the face of challenges
# Interpersonal sensitivity: Measure of sensitivity and awareness of others' needs
```

```python
# Check if there are any missing values in the entire dataset    
data.isna().any().any()

# Remove row with missing values
data = data.dropna()
```

```python
# Transpose the DataFrame
data = data.transpose()
display(data.head())
#rows from 1 represent individual respondents while columns represent the names of items scored within the survey
```

```python
# Set row 1 as the variable names
data = data.set_axis(data.iloc[0], axis=1)
data = data[1:]  # Exclude row 1 from the data
```

```python
#Descriptive statistics
data.mean()
data.std()
```

```python
#HYPOTHESIS FORMULATION

#Null Hypothesis (H0): There are no distinct entrepreneurial mindset profiles or clusters present.

#Alternative Hypothesis (Ha): Different entrepreneurial mindset profiles or clusters can be identified.

#Hypothesis explanation

#The null hypothesis assumes that there are no discernible patterns or clusters in the entrepreneurial mindsets of the respondents based on the provided data. On the other hand, the alternative hypothesis suggests that there are indeed distinct profiles or clusters of entrepreneurial mindsets that can be identified in the dataset.
```

```python
#Data visualization

#Boxplot

plt.figure(figsize=(26,6))
sns.boxplot(data=data )
plt.title('Measure of Skill and Personality Traits')
plt.ylabel('Measure')
plt.xlabel('Skill and Personality Traits')
plt.show()
```

```python
#Barplot
plt.figure(figsize=(26,8))
sns.barplot(data=data )
plt.title('Measure of Skill and Personality Traits')
plt.ylabel('Measure')
plt.xlabel('Skill and Personality Traits')
plt.xticks(rotation=20)
plt.show()

#From the plot, most respondents acknowledged the need to have high risk acceptance, self confidence, perserverance, and perfomance need for the success of a business.
```

```python
#Cluster analysis

# Select the relevant columns for clustering
columns = ['independence ', 'structure ', 'non-confromity', 'risk acceptance ', 'Action orientation', 
           'Passion', 'Performance need', 'Focus on the future', 'Idea generation', 
           'export', 'Self confidence', 'Optimism', 'Perseverance', 'Interpersonal sensitivity', 'Cluster']
cluster_data = data[columns]

# Choose the number of clusters
num_clusters = 3  # Adjust the number of clusters based on your needs

# Perform K-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(cluster_data)

# Get the cluster labels
cluster_labels = kmeans.labels_

# Add the cluster labels to the original dataset
data['Cluster'] = cluster_labels

# View the cluster assignments
print(data[['Cluster'] + columns])

#Interpretation 

#Based on the provided output, the dataset contains information on respondents' scores for various attributes related to entrepreneurial mindset. The attributes include independence, structure, non-conformity, risk acceptance, action orientation, passion, performance need, focus on the future, idea generation, export, self-confidence, optimism, perseverance, and interpersonal sensitivity.

#The output also includes a "Cluster" column, which indicates the assigned cluster for each respondent based on the cluster analysis.

#To interpret the output, we can analyze the characteristics of each cluster by examining the mean scores of the attributes within each cluster.

#Cluster 0:

#Respondents in this cluster tend to have relatively lower scores in independence, structure, risk acceptance, and focus on the future compared to the other clusters.
#They have moderate scores in non-conformity, action orientation, passion, performance need, idea generation, and perseverance.
#They have relatively higher scores in self-confidence, optimism, and interpersonal sensitivity.

#Cluster 1:

#Respondents in this cluster have moderate scores across most attributes.
#They have relatively higher scores in structure, non-conformity, action orientation, passion, performance need, and idea generation.
#They have lower scores in focus on the future and self-confidence.

#Cluster 2:

#Respondents in this cluster tend to have higher scores in most attributes compared to the other clusters.
#They have relatively higher scores in independence, structure, risk acceptance, action orientation, passion, performance need, focus on the future, idea generation, and perseverance.
#They have moderate scores in non-conformity, self-confidence, optimism, and interpersonal sensitivity.
```
